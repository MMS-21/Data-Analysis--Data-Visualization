## SETUP AND IMPORTS

```python
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import seaborn as sns
import sqlite3
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Get project root (parent of this notebook's folder)
project_root = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
sys.path.append(project_root)
# Import churn functions
from src.churn import *
from src.analysis_utils import print_business_summary

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 14


# Connect to database
DB_PATH = '../db/retail.db'
conn = sqlite3.connect(DB_PATH)

print("‚úÖ Libraries and functions loaded successfully")
```


## 1. LOAD AND PREPARE DATA

```python
# Load transactions
df = pd.read_sql("SELECT * FROM transactions", conn)
df['invoice_date'] = pd.to_datetime(df['invoice_date'])

print(f"Dataset: {len(df):,} transactions")
print(f"Customers: {df['customer_id'].nunique():,}")
print(f"Date Range: {df['invoice_date'].min().date()} to {df['invoice_date'].max().date()}")

# Business summary
print_business_summary(df)
```

## 2. DEFINE CHURN

```python
print("\n\n" + "="*80)
print(" DEFINING CHURN")
print("="*80)

# Define churn threshold (90 days without purchase)
CHURN_THRESHOLD_DAYS = 90

churn_labels = define_churn(df, churn_threshold_days=CHURN_THRESHOLD_DAYS)

print(f"\nChurn Threshold: {CHURN_THRESHOLD_DAYS} days")
print(f"\nChurn Distribution:")
print(churn_labels['is_churned'].value_counts())
print(f"\nChurn Rate: {churn_labels['is_churned'].mean()*100:.2f}%")

# Visualize churn distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Churn counts
churn_counts = churn_labels['is_churned'].value_counts()
axes[0].bar(['Active', 'Churned'], churn_counts.values, color=['#2E86AB', '#A23B72'], alpha=0.8)
axes[0].set_title('Customer Churn Distribution', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Number of Customers')
for i, v in enumerate(churn_counts.values):
    axes[0].text(i, v + 50, f'{v:,}', ha='center', fontweight='bold')

# Churn pie chart
axes[1].pie(churn_counts.values, labels=['Active', 'Churned'], autopct='%1.1f%%',
            colors=['#2E86AB', '#A23B72'], startangle=90)
axes[1].set_title('Churn Rate', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

## 3. FEATURE ENGINEERING

```python
print("\n" + "="*80)
print(" FEATURE ENGINEERING")
print("="*80)

# Engineer features
features_df = engineer_churn_features(df, churn_labels)

print(f"\nFeatures created for {len(features_df):,} customers")
print(f"Number of features: {len(features_df.columns) - 2}")  # Exclude customer_id and target

print("\nüìä Sample Features:")
print(features_df.head())

print("\nüìà Feature Statistics:")
feature_cols = [col for col in features_df.columns if col not in ['customer_id', 'is_churned', 'country_clean']]
print(features_df[feature_cols].describe())
```

## 4. EXPLORATORY ANALYSIS OF FEATURES

```python
print("\n" + "="*80)
print(" FEATURE ANALYSIS BY CHURN STATUS")
print("="*80)

# Compare churned vs active customers
churned = features_df[features_df['is_churned'] == 1]
active = features_df[features_df['is_churned'] == 0]

comparison_metrics = {
    'Recency (days)': ('recency_days', churned['recency_days'].mean(), active['recency_days'].mean()),
    'Frequency': ('frequency', churned['frequency'].mean(), active['frequency'].mean()),
    'Monetary ($)': ('monetary', churned['monetary'].mean(), active['monetary'].mean()),
    'Avg Order Value ($)': ('avg_order_value', churned['avg_order_value'].mean(), active['avg_order_value'].mean()),
    'Customer Age (days)': ('customer_age_days', churned['customer_age_days'].mean(), active['customer_age_days'].mean()),
    'Unique Products': ('unique_products', churned['unique_products'].mean(), active['unique_products'].mean())
}

print("\n Churned vs Active Customers:")
print(f"{'Metric':<25} {'Churned':>15} {'Active':>15} {'Difference':>15}")
print("-" * 75)

for metric_name, (col, churned_val, active_val) in comparison_metrics.items():
    diff = churned_val - active_val
    print(f"{metric_name:<25} {churned_val:>15.2f} {active_val:>15.2f} {diff:>15.2f}")

# Visualize key differences
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

plot_features = ['recency_days', 'frequency', 'monetary', 
                 'avg_order_value', 'customer_age_days', 'unique_products']
titles = ['Recency', 'Frequency', 'Monetary', 
          'Avg Order Value', 'Customer Age', 'Unique Products']

for idx, (feature, title) in enumerate(zip(plot_features, titles)):
    axes[idx].hist(active[feature], bins=30, alpha=0.6, label='Active', color='#2E86AB')
    axes[idx].hist(churned[feature], bins=30, alpha=0.6, label='Churned', color='#A23B72')
    axes[idx].set_title(title, fontweight='bold')
    axes[idx].set_xlabel(title)
    axes[idx].set_ylabel('Count')
    axes[idx].legend()

plt.tight_layout()
plt.show()
```

## 5. PREPARE DATA FOR MODELING

```python
print("\n" + "="*80)
print(" PREPARING DATA FOR MACHINE LEARNING")
print("="*80)

# Prepare model data
X_train, X_test, y_train, y_test, feature_names, scaler = prepare_model_data(features_df)

print(f"\nTraining set: {len(X_train):,} customers")
print(f"Test set: {len(X_test):,} customers")
print(f"Number of features: {len(feature_names)}")

print(f"\nClass distribution in training set:")
print(y_train.value_counts())
print(f"Churn rate: {y_train.mean()*100:.2f}%")

print(f"\nClass distribution in test set:")
print(y_test.value_counts())
print(f"Churn rate: {y_test.mean()*100:.2f}%")
```

# 6. TRAIN MODELS

```python
print("\n" + "="*80)
print(" TRAINING MACHINE LEARNING MODELS")
print("="*80)

# Train Logistic Regression
print("\n1Ô∏è‚É£  Training Logistic Regression...")
lr_model, lr_pred, lr_prob, lr_metrics = train_logistic_regression(X_train, y_train, X_test, y_test)
print(f"   ‚úÖ Accuracy: {lr_metrics['accuracy']:.4f}")
print(f"   ‚úÖ ROC-AUC: {lr_metrics['roc_auc']:.4f}")
print(f"   ‚úÖ F1-Score: {lr_metrics['f1_score']:.4f}")

# Train Random Forest
print("\n2Ô∏è‚É£  Training Random Forest...")
rf_model, rf_pred, rf_prob, rf_metrics = train_random_forest(X_train, y_train, X_test, y_test)
print(f"   ‚úÖ Accuracy: {rf_metrics['accuracy']:.4f}")
print(f"   ‚úÖ ROC-AUC: {rf_metrics['roc_auc']:.4f}")
print(f"   ‚úÖ F1-Score: {rf_metrics['f1_score']:.4f}")

# Train Gradient Boosting
print("\n3Ô∏è‚É£  Training Gradient Boosting...")
gb_model, gb_pred, gb_prob, gb_metrics = train_gradient_boosting(X_train, y_train, X_test, y_test)
print(f"   ‚úÖ Accuracy: {gb_metrics['accuracy']:.4f}")
print(f"   ‚úÖ ROC-AUC: {gb_metrics['roc_auc']:.4f}")
print(f"   ‚úÖ F1-Score: {gb_metrics['f1_score']:.4f}")
```

## 7. COMPARE MODELS

```python

print(" MODEL COMPARISON")
print("="*80)

models_dict = {
    'Logistic Regression': (lr_model, lr_pred, lr_prob, lr_metrics),
    'Random Forest': (rf_model, rf_pred, rf_prob, rf_metrics),
    'Gradient Boosting': (gb_model, gb_pred, gb_prob, gb_metrics)
}

comparison_df = compare_models(models_dict, X_test, y_test)
print("\nüèÜ Model Performance Comparison:")
print(comparison_df.to_string(index=False))

# Select best model
best_model_name = comparison_df.iloc[0]['Model']
best_model, best_pred, best_prob, best_metrics = models_dict[best_model_name]

print(f"\n Best Model: {best_model_name}")
print(f"   ‚Ä¢ ROC-AUC: {best_metrics['roc_auc']:.4f}")
print(f"   ‚Ä¢ F1-Score: {best_metrics['f1_score']:.4f}")

# Visualize comparison
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(comparison_df))
width = 0.25

ax.bar(x - width, comparison_df['Accuracy'], width, label='Accuracy', color='#2E86AB', alpha=0.8)
ax.bar(x, comparison_df['ROC-AUC'], width, label='ROC-AUC', color='#A23B72', alpha=0.8)
ax.bar(x + width, comparison_df['F1-Score'], width, label='F1-Score', color='#F18F01', alpha=0.8)

ax.set_xlabel('Model', fontsize=12)
ax.set_ylabel('Score', fontsize=12)
ax.set_title('Model Performance Comparison', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(comparison_df['Model'])
ax.legend()
ax.set_ylim(0, 1.0)

plt.tight_layout()
plt.show()
```

## 8. DETAILED EVALUATION OF BEST MODEL

```python

print(f" DETAILED EVALUATION - {best_model_name}")


# Classification report
evaluate_classification_report(y_test, best_pred, best_model_name)

# Confusion matrix
print("\n Confusion Matrix:")
plot_confusion_matrix(y_test, best_pred, best_model_name)

# ROC Curve
print("\n ROC Curve:")
plot_roc_curve(y_test, best_prob, best_model_name)

# Precision-Recall Curve
print("\n Precision-Recall Curve:")
plot_precision_recall_curve(y_test, best_prob, best_model_name)
```

## 9. FEATURE IMPORTANCE

```python

print(" FEATURE IMPORTANCE ANALYSIS")
print("="*80)

if best_model_name in ['Random Forest', 'Gradient Boosting']:
    print(f"\n Top Features for Predicting Churn ({best_model_name}):")
    feature_importance = plot_feature_importance(best_model, feature_names, top_n=15)
    print("\nTop 10 Most Important Features:")
    print(feature_importance.head(10).to_string(index=False))
else:
    print(f"\n‚ö†Ô∏è  Feature importance not available for {best_model_name}")
    print("Using coefficient magnitude instead:")
    
    # For logistic regression, use coefficients
    coef_df = pd.DataFrame({
        'feature': feature_names,
        'coefficient': np.abs(lr_model.coef_[0])
    }).sort_values('coefficient', ascending=False)
    
    print("\nTop 10 Features by Coefficient Magnitude:")
    print(coef_df.head(10).to_string(index=False))
    
    # Plot
    top_coef = coef_df.head(15)
    plt.figure(figsize=(10, 6))
    plt.barh(range(len(top_coef)), top_coef['coefficient'], color='#2E86AB', alpha=0.8)
    plt.yticks(range(len(top_coef)), top_coef['feature'].to_list())
    plt.xlabel('|Coefficient|', fontsize=12)
    plt.title('Top 15 Features by Coefficient Magnitude', fontsize=14, fontweight='bold')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```

## 10. CALCULATE CHURN RISK SCORES FOR ALL CUSTOMERS

```python
print("\n" + "="*80)
print(" CALCULATING CHURN RISK SCORES FOR ALL CUSTOMERS")
print("="*80)

# Prepare all customer data
X_all = features_df.drop(['customer_id', 'is_churned'], axis=1)
X_all_encoded = pd.get_dummies(X_all, columns=['country_clean'], drop_first=True)

# Ensure same columns as training data
missing_cols = set(feature_names) - set(X_all_encoded.columns)
for col in missing_cols:
    X_all_encoded[col] = 0
X_all_encoded = X_all_encoded[feature_names]

# Calculate risk scores
risk_scores = calculate_churn_risk_scores(best_model, features_df, X_all_encoded, scaler)

print(f"\nChurn risk calculated for {len(risk_scores):,} customers")
print(f"\n Risk Category Distribution:")
print(risk_scores['risk_category'].value_counts().sort_index())

# Visualize risk distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Risk category counts
risk_counts = risk_scores['risk_category'].value_counts().sort_index()
axes[0].bar(risk_counts.index, risk_counts.values, 
           color=['#2E86AB', '#F18F01', '#D62828', '#9B2226'], alpha=0.8)
axes[0].set_title('Customers by Risk Category', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Number of Customers')
axes[0].tick_params(axis='x', rotation=45)

# Churn probability distribution
axes[1].hist(risk_scores['churn_probability'], bins=50, color='#A23B72', alpha=0.7, edgecolor='black')
axes[1].set_title('Churn Probability Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Churn Probability')
axes[1].set_ylabel('Number of Customers')
axes[1].axvline(0.5, color='red', linestyle='--', label='50% Threshold')
axes[1].legend()

plt.tight_layout()
plt.show()
```

## 11. IDENTIFY HIGH-RISK CUSTOMERS

```python
print("\n" + "="*80)
print(" HIGH-RISK CUSTOMERS (Immediate Action Required)")
print("="*80)

# Get customer profiles with risk
customer_profiles = segment_by_churn_risk(risk_scores, features_df)

# High risk customers
high_risk = customer_profiles[customer_profiles['risk_category'].isin(['High Risk', 'Critical Risk'])]

print(f"\nHigh & Critical Risk Customers: {len(high_risk):,}")
print(f"Percentage of total: {len(high_risk)/len(customer_profiles)*100:.1f}%")
print(f"Revenue at Risk: ${high_risk['monetary'].sum():,.2f}")

# Show top at-risk customers
print("\n HIGH-RISK CUSTOMERS (Immediate Action Required)")
top_risk = high_risk.nlargest(20, 'monetary')
print(top_risk[['customer_id', 'churn_probability', 'risk_category', 
                'recency_days', 'frequency', 'monetary', 'avg_order_value']].to_string(index=False))
```

## 12. COMBINE WITH RFM SEGMENTS (If available)

```python
print("\n" + "="*80)
print("üîÑ INTEGRATING CHURN RISK WITH CUSTOMER SEGMENTS")
print("="*80)

# Try to load RFM data if it exists
try:
    rfm_data = pd.read_csv('../reports/rfm_clv_analysis.csv')
    
    # Merge churn risk with RFM segments
    integrated = customer_profiles.merge(
        rfm_data[['customer_id', 'segment', 'value_segment', 'predicted_clv']], 
        on='customer_id', 
        how='left'
    )
    
    print("\n‚úÖ Integrated churn risk with RFM segments")
    
    # Analyze churn risk by segment
    print("\nüìä Churn Risk by RFM Segment:")
    segment_churn = integrated.groupby('segment').agg({
        'customer_id': 'count',
        'churn_probability': 'mean',
        'monetary': 'sum'
    }).reset_index()
    
    segment_churn.columns = ['segment', 'num_customers', 'avg_churn_prob', 'total_revenue']
    segment_churn = segment_churn.sort_values('avg_churn_prob', ascending=False)
    
    print(segment_churn.to_string(index=False))
    
    # High-value customers at risk
    high_value_at_risk = integrated[
        (integrated['risk_category'].isin(['High Risk', 'Critical Risk'])) &
        (integrated['value_segment'] == 'High Value')
    ]
    
    print(f"\nüö® High-Value Customers at Risk of Churn: {len(high_value_at_risk):,}")
    print(f"   Revenue at Risk: ${high_value_at_risk['monetary'].sum():,.2f}")
    
except FileNotFoundError:
    print("\n‚ö†Ô∏è  RFM analysis file not found. Run 02-rfm-clv.ipynb first for integrated insights.")
    integrated = customer_profiles.copy()
```
